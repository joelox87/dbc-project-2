```python
import pandas as pd
```

```python
csv_path = 'genres.csv'

genres_df = pd.read_csv(csv_path, encoding="utf-8")
genres_df.head()
```

```python
# delete columns that aren't neccessary to display

del genres_df['mode'] 
del genres_df['acousticness']
del genres_df['danceability']
del genres_df['duration_ms']
del genres_df['energy']
del genres_df['instrumentalness']
del genres_df['liveness'] 
del genres_df['loudness']
del genres_df['speechiness']
del genres_df['tempo']
del genres_df['valence']
del genres_df['key']

genres_df.head()
```

```python
#sort popularity column by most to least popular

genres_df.sort_values(by='popularity', ascending=False)
```

```python
#show top 10 and bottom 10 in popularity row

genres_df.head(10).append(genres_df.tail(10))
```

```python
genres_df.loc[(genres_df!=0.0).any(1)]

genres_df.tail(10)
```

```python
#get rid of rows that have a value of 0

genres_df= genres_df[genres_df['popularity'] != 0]

genres_df.head(10).append(genres_df.tail(10))
```
