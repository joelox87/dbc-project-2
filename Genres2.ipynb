```python
# Import SQLAlchemy `automap` and other dependencies here
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func

database_path = "../DBC-PROJECT-2/spotify_genres.sqlite"
```

```python
import pandas as pd
```

```python
#Create an engine for the `spotify.sqlite` database
engine = create_engine(f"sqlite:///{database_path}", echo=False)
conn = engine.connect()
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
```

```python
csv_path = 'genres.csv'

genres_df = pd.read_csv(csv_path, encoding="utf-8")
genres_df.head()
```

```python
#sort popularity column by most to least popular

genres_sorted_df = genres_df.sort_values(by='popularity', ascending=False)
genres_sorted_df
```

```python
#show top 10 and bottom 10 in popularity row

genres_sorted_df.head(100).append(genres_sorted_df.tail(100))
```

```python
genres_sorted_df.loc[(genres_sorted_df!=0.0).any(1)]

genres_sorted_df.tail(100)
```

```python
#get rid of rows that have a value of 0

genres_df_filtered= genres_sorted_df[genres_sorted_df['popularity'] != 0]

df_final = genres_df_filtered.head(100).append(genres_df_filtered.tail(100))
df_final
```

```python
df_final = df_final.reset_index(drop = True)
df_final
```

```python
df_final.to_sql("sortedgenres",conn)
```
